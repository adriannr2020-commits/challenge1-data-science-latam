### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
TTL_tienda1 = sum(tienda["Precio"])
TTL_tienda2 = sum(tienda2["Precio"])
TTL_tienda3 = sum(tienda3["Precio"])
TTL_tienda4 = sum(tienda4["Precio"])

print(
    f"El total de ventas de cada tienda es:\n"
    f"TIENDA 1 = {TTL_tienda1}\n"
    f"TIENDA 2 = {TTL_tienda2}\n"
    f"TIENDA 3 = {TTL_tienda3}\n"
    f"TIENDA 4 = {TTL_tienda4}"
     )
print("La tienda con menor facturacion es la N° 4")
```

# 2. Ventas por categoría

```python
top_prods = sum([
            tienda["Categoría del Producto"].value_counts(),
            tienda2["Categoría del Producto"].value_counts(),
            tienda3["Categoría del Producto"].value_counts(),
            tienda4["Categoría del Producto"].value_counts()
                ])

print(top_prods.sort_values(ascending=False))
```

```python
cant_prod_1 = tienda["Categoría del Producto"].value_counts()
cant_prod_2 = tienda2["Categoría del Producto"].value_counts()
cant_prod_3 = tienda3["Categoría del Producto"].value_counts()
cant_prod_4 = tienda4["Categoría del Producto"].value_counts()

matriz = pd.DataFrame({
    "TIENDA 1": cant_prod_1,
    "TIENDA 2": cant_prod_2,
    "TIENDA 3": cant_prod_3,
    "TIENDA 4": cant_prod_4
}).fillna(0)

matriz.loc["TOTALES"] = matriz.sum(axis=0)

matriz = matriz[matriz.loc["TOTALES"].sort_values(ascending=False).index]

print(matriz)
```

```python
top_prods = sum([cant_prod_1, cant_prod_2, cant_prod_3, cant_prod_4])

top3 = top_prods.sort_values(ascending=False).head(3)
top3.index.name = None
top3.name = None

print("Los 3 productos más vendidos en las 4 tiendas son:\n"
      f"{top3.to_string()}")

print(f"\nY la tienda con menor cantidad de productos es la TIENDA 4 con {sum(cant_prod_4)} unidades entre todas las categorias")
```

# 3. Calificación promedio de la tienda


```python
calif_1 = tienda["Calificación"].value_counts()
calif_2 = tienda2["Calificación"].value_counts()
calif_3 = tienda3["Calificación"].value_counts()
calif_4 = tienda4["Calificación"].value_counts()

matriz_calif = pd.concat([calif_1, calif_2, calif_3, calif_4], axis=1)
matriz_calif = matriz_calif.fillna(0)
matriz_calif.columns = ["TIENDA 1", "TIENDA 2", "TIENDA 3", "TIENDA 4"]

matriz_calif.loc["TOTALES"] = matriz_calif.sum(axis=0)

matriz_calif = matriz_calif[matriz_calif.loc["TOTALES"].sort_values(ascending=False).index]

print(matriz_calif)
```

```python
prom1 = tienda["Calificación"].mean()
prom2 = tienda2["Calificación"].mean()
prom3 = tienda3["Calificación"].mean()
prom4 = tienda4["Calificación"].mean()

df_promedios = pd.DataFrame({
    "Tienda": ["TIENDA 1", "TIENDA 2", "TIENDA 3", "TIENDA 4"],
    "Calif. promedio ": [prom1, prom2, prom3, prom4]
                            })
print(df_promedios.to_string(index=False))
print(f"La tienda con menor calificacion es la TIENDA 1 con {prom1:.3f}")
```

#4. Productos más y menos vendidos

```python
top_1 = tienda["Producto"].value_counts().head(5)
top_2 = tienda2["Producto"].value_counts().head(5)
top_3 = tienda3["Producto"].value_counts().head(5)
top_4 = tienda4["Producto"].value_counts().head(5)

top_2
```

```python
df_prods_t1 = tienda.groupby("Producto").size().reset_index(name='Cant.ventas').sort_values(by='Cant.ventas', ascending=False).reset_index(drop=True)
df_prods_t2 = tienda2.groupby("Producto").size().reset_index(name='Cant.ventas').sort_values(by='Cant.ventas', ascending=False).reset_index(drop=True)
df_prods_t3 = tienda3.groupby("Producto").size().reset_index(name='Cant.ventas').sort_values(by='Cant.ventas', ascending=False).reset_index(drop=True)
df_prods_t4 = tienda4.groupby("Producto").size().reset_index(name='Cant.ventas').sort_values(by='Cant.ventas', ascending=False).reset_index(drop=True)

df_prods_t1["Pto. de Venta"] = "TIENDA1"
df_prods_t2["Pto. de Venta"] = "TIENDA2"
df_prods_t3["Pto. de Venta"] = "TIENDA3"
df_prods_t4["Pto. de Venta"] = "TIENDA4"

df_prods_t1 = df_prods_t1[["Pto. de Venta", "Producto", "Cant.ventas"]]
df_prods_t2 = df_prods_t2[["Pto. de Venta", "Producto", "Cant.ventas"]]
df_prods_t3 = df_prods_t3[["Pto. de Venta", "Producto", "Cant.ventas"]]
df_prods_t4 = df_prods_t4[["Pto. de Venta", "Producto", "Cant.ventas"]]

max_filas = max(len(df_prods_t1), len(df_prods_t2), len(df_prods_t3), len(df_prods_t4))

df_prods_t1 = df_prods_t1.reindex(range(max_filas))
df_prods_t2 = df_prods_t2.reindex(range(max_filas))
df_prods_t3 = df_prods_t3.reindex(range(max_filas))
df_prods_t4 = df_prods_t4.reindex(range(max_filas))

df_matriz_todo = pd.concat([df_prods_t1, df_prods_t2, df_prods_t3, df_prods_t4], axis=1)

df_matriz_todo
```

```python
print("Cantidad de productos vendidos en las 4 tiendas (del mas al menos vendido):\n")
df_matriz_todo
```

```python
print("Los 5 productos MÁS VENDIDOS de las 4 tiendas:\n")
print(df_matriz_todo.head(5))

print("\nLos 5 productos MENOS VENDIDOS de las 4 tiendas:\n")
print(df_matriz_todo.tail(5))
```

```python
from IPython.display import display

print("Los 5 productos MÁS VENDIDOS de las 4 tiendas son:\n")
display(df_matriz_todo.head(5))

print("\nY los 5 productos MENOS VENDIDOS, son:\n")
display(df_matriz_todo.tail(5))
```

```python
import matplotlib.pyplot as plt

top_n = 5
top_t1 = df_prods_t1.head(top_n)
top_t2 = df_prods_t2.head(top_n)
top_t3 = df_prods_t3.head(top_n)
top_t4 = df_prods_t4.head(top_n)

fig, axes = plt.subplots(2, 2, figsize=(14,5))

tiendas = [top_t1, top_t2, top_t3, top_t4]
colores = ['skyblue','lightgreen','salmon','orange']
titulos = ['TIENDA 1','TIENDA 2','TIENDA 3','TIENDA 4']

for ax, df, color, titulo in zip(axes.flatten(), tiendas, colores, titulos):
    ax.barh(df["Producto"], df["Cant.ventas"], color=color)
    ax.set_title(titulo)
    ax.set_xlabel("Cantidad de Ventas")
    ax.set_ylabel("Producto")
    ax.invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
mejores = 5
peores = 5

mejores_t1 = df_prods_t1.head(mejores)
peores_t1 = df_prods_t1.tail(peores)

mejores_t2 = df_prods_t2.head(mejores)
peores_t2 = df_prods_t2.tail(peores)

mejores_t3 = df_prods_t3.head(mejores)
peores_t3 = df_prods_t3.tail(peores)

mejores_t4 = df_prods_t4.head(mejores)
peores_t4 = df_prods_t4.tail(peores)

fig, axes = plt.subplots(4, 2, figsize=(16, 20))

mejores_tienda_n = [mejores_t1, mejores_t2, mejores_t3, mejores_t4]
peores_tienda_n = [peores_t1, peores_t2, peores_t3, peores_t4]
color_mejor_n = ['skyblue','lightgreen','salmon','orange']
color_peor_n = ['blue','green','red','darkorange']
titulos = ['TIENDA 1','TIENDA 2','TIENDA 3','TIENDA 4']

for i in range(4):
    axes[i,0].barh(mejores_tienda_n[i]["Producto"], mejores_tienda_n[i]["Cant.ventas"], color=color_mejor_n[i])
    axes[i,0].set_title(f"{titulos[i]} - LOS {mejores} PRODUCTOS MAS VENDIDOS ")
    axes[i,0].invert_yaxis()
    axes[i,0].set_xlabel("Cantidad de Ventas")

    axes[i,1].barh(peores_tienda_n[i]["Producto"], peores_tienda_n[i]["Cant.ventas"], color=color_peor_n[i])
    axes[i,1].set_title(f"{titulos[i]} - LOS {peores} PRODUCTOS MENOS VENDIDOS ")
    axes[i,1].invert_yaxis()
    axes[i,1].set_xlabel("Cantidad de Ventas")

plt.tight_layout()
plt.show()
```
