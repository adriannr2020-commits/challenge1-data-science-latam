# 2. Ventas por categoría

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
top_prods = None

for tienda_df in tiendas:
    if top_prods is None:
        top_prods = tienda_df["Categoría del Producto"].value_counts()
    else:
        top_prods = top_prods + tienda_df["Categoría del Producto"].value_counts()

print(top_prods.sort_values(ascending=False))
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
datos = {}
i = 1

for tienda_df in tiendas:
    datos[f"TIENDA {i}"] = tienda_df["Categoría del Producto"].value_counts()
    i += 1

matriz = pd.DataFrame(datos)

matriz.loc["TOTALES"] = matriz.sum(axis=0)

matriz = matriz[matriz.loc["TOTALES"].sort_values(ascending=False).index]

print(matriz)
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
top_prods = None

for tienda_df in tiendas:
    if top_prods is None:
        top_prods = tienda_df["Categoría del Producto"].value_counts()
    else:
        top_prods = top_prods + tienda_df["Categoría del Producto"].value_counts()

top3 = top_prods.sort_values(ascending=False).head(3)
top3.index.name = None

print("Los 3 productos más vendidos en las 4 tiendas son:\n"
      f"{top3.to_string()}")

print(f"\nY la tienda con menor cantidad de productos vendidos es la TIENDA 4 con {tienda4['Categoría del Producto'].value_counts().sum()} unidades entre todas las categorias")
```

# 3. Calificación promedio de la tienda


```python
tiendas = [tienda, tienda2, tienda3, tienda4]
califaciones = []

for tienda_df in tiendas:
    califaciones.append(tienda_df["Calificación"].value_counts())

matriz_calif = pd.concat(califaciones, axis=1)
matriz_calif = matriz_calif.fillna(0)
matriz_calif.columns = ["TIENDA 1", "TIENDA 2", "TIENDA 3", "TIENDA 4"]

print(matriz_calif)
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
promedios = []

for i, tienda_df in enumerate(tiendas, 1):
    promedios.append(tienda_df["Calificación"].mean())

df_promedios = pd.DataFrame({
    "Tienda": ["TIENDA 1", "TIENDA 2", "TIENDA 3", "TIENDA 4"],
    "Calif. promedio ": promedios
})

print(df_promedios.to_string(index=False))
print(f"La tienda con menor calificacion es la TIENDA 1 con {promedios[0]:.3f}")
```

```python
print("ANÁLISIS DE VENTAS POR TIENDA:\n")

tiendas = [tienda, tienda2, tienda3, tienda4]
i = 1

for tienda_df in tiendas:
    df_prods = tienda_df.groupby("Producto").size().reset_index(name='Cant.vtas.').sort_values(by='Cant.vtas.', ascending=False).reset_index(drop=True)
    print(f"\nTIENDA {i}")
    print(df_prods[["Producto", "Cant.vtas."]].to_string(index=False))
    i += 1
```

```python
tiendas = [tienda, tienda2, tienda3, tienda4]
dfs_prods = []
i = 1

for tienda_df in tiendas:
    df_prods = tienda_df.groupby("Producto").size().reset_index(name='Cant.vtas.').sort_values(by='Cant.vtas.', ascending=False).reset_index(drop=True)
    df_prods = df_prods[["Producto", "Cant.vtas."]]
    df_prods.columns = [f"TIENDA {i}", f"Cant.vtas. {i}"]
    dfs_prods.append(df_prods)
    i += 1

df_final = pd.concat(dfs_prods, axis=1)

print("CANTIDAD DE PRODUCTOS VENDIDOS EN LAS 4 TIENDAS (del mas al menos vendido):\n")
print(df_final.to_string(index=False))
```

```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]
dfs_prods = []
i = 1

for tienda_df in tiendas:
    df_prods = tienda_df.groupby("Producto").size().reset_index(name='Cant.ventas').sort_values(by='Cant.ventas', ascending=False).reset_index(drop=True)
    df_prods = df_prods[["Producto", "Cant.ventas"]]
    dfs_prods.append(df_prods)
    i += 1

top_n = 5
colores = ['skyblue','lightgreen','salmon','orange']
titulos = ['TIENDA 1','TIENDA 2','TIENDA 3','TIENDA 4']

fig, axes = plt.subplots(2, 2, figsize=(14, 5))
axes = axes.flatten()

for i in range(len(dfs_prods)):
    ax = axes[i]
    df = dfs_prods[i]
    color = colores[i]
    titulo = titulos[i]

    ax.barh(df.head(top_n)["Producto"], df.head(top_n)["Cant.ventas"], color=color)
    ax.set_title(titulo)
    ax.set_xlabel("Cantidad de Ventas")
    ax.set_ylabel("Producto")
    ax.invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
print("\n ANALISIS DE RENDIMIENTO POR TIENDA:\n")

tiendas = [tienda, tienda2, tienda3, tienda4]
datos_rendimiento = []
i = 1

for tienda_df in tiendas:
    total_ventas = tienda_df["Precio"].sum()
    cantidad_productos = len(tienda_df)
    promedio_ticket = total_ventas / cantidad_productos

    datos_rendimiento.append({
        "Tienda": f"TIENDA {i}",
        "Total de ventas": f"${total_ventas:.2f}",
        "Cant. de productos": cantidad_productos,
        "Ticket promedio": f"${promedio_ticket:.2f}"
    })

    i += 1
df_rendimiento = pd.DataFrame(datos_rendimiento)

print(df_rendimiento.to_string(index=False))

print(f"\nLa tienda con MENOR RENDIMIENTO es la TIENDA 4")
```

# 5. Envío promedio por tienda

```python
print("ANÁLISIS DE POSIBLES DATOS FALTANTES Y ENVÍOS SIN COSTO:\n")

tiendas = [tienda, tienda2, tienda3, tienda4]
datos_faltantes_list = []
i = 1

for tienda_df in tiendas:
    datos_faltantes = tienda_df["Costo de envío"].isna().sum()
    envios_cero = (tienda_df["Costo de envío"] == 0).sum()
    total_registros = len(tienda_df)
    registros_validos = total_registros - datos_faltantes - envios_cero

    datos_faltantes_list.append({
        "Tienda": f"TIENDA {i}",
        "Total de registros": total_registros,
        "Sin datos": datos_faltantes,
        "Envíos a costo 0": envios_cero,
        "Registros válidos": registros_validos
    })

    i += 1

df_datos_faltantes = pd.DataFrame(datos_faltantes_list)

print(df_datos_faltantes.to_string(index=False))

print(f"\nLa cantidad de envíos sin costo no es significativamente alta en ninguna de las tiendas.\nEl nro. de registros validos tomados en la metrica, practicamente no varia de una tienda a otra.")
```

```python
print("ANÁLISIS DE COSTO DE ENVÍO PROMEDIO POR TIENDA:\n")

tiendas = [tienda, tienda2, tienda3, tienda4]
datos_envio = []
i = 1

for tienda_df in tiendas:
    costo_envio_promedio = tienda_df["Costo de envío"].mean()
    costo_envio_total = tienda_df["Costo de envío"].sum()
    cantidad_envios = len(tienda_df)

    datos_envio.append({
        "Tienda": f"TIENDA {i}",
        "Cantidad de envíos": cantidad_envios,
        "$ Total envíos": f"${costo_envio_total:.2f}",
        "$ Promedio envíos": f"${costo_envio_promedio:.2f}"
    })

    i += 1

df_envio = pd.DataFrame(datos_envio)

print(df_envio.to_string(index=False))

print(f"\nLa tienda con menor VALOR PROMEDIO DE ENVIOS es la TIENDA 4")
```

```python
print("ANÁLISIS DE MARGEN DE RENDIMIENTO PARA TOMA DE DECISION FINAL:\n")

tiendas = [tienda, tienda2, tienda3, tienda4]
datos_margen = []
i = 1

for tienda_df in tiendas:
    total_ventas = tienda_df["Precio"].sum()
    total_envios = tienda_df["Costo de envío"].sum()
    ganancia_neta = total_ventas - total_envios
    cantidad_productos = len(tienda_df)
    promedio_ticket = total_ventas / cantidad_productos
    margen_ganancia = (ganancia_neta / total_ventas * 100) if total_ventas > 0 else 0
    envios_sin_costo = (tienda_df["Costo de envío"] == 0).sum()
    costo_envio_promedio = tienda_df["Costo de envío"].mean()

    datos_margen.append({
        "Tienda": f"TIENDA {i}",
        "Total vtas.": f"${total_ventas:.0f}",
        "Total envíos": f"${total_envios:.0f}",
        "Ganancia neta": f"${ganancia_neta:.0f}",
        "Margen de ganancia": f"{margen_ganancia:.2f}%",
        "Ticket promedio": f"${promedio_ticket:.2f}",
        "Envíos sin costo": envios_sin_costo,
        "Cant. de transacciones": cantidad_productos,
        "Costo prom. envíos": f"${costo_envio_promedio:.2f}"
    })

    i += 1

df_margen = pd.DataFrame(datos_margen)

print(df_margen.to_string(index=False))

print("\nSi bien es llamativo que el margen de ganacia es el mismo en todas las tiendas, luego de analizar ventas y envíos\nla tienda que presenta menor margen de ganancia y menor ticket promedio es la TIENDA 4, por lo cual sería la tienda a cerrar.")
```
